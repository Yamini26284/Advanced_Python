```python
""" Read about explicit and implicit type casting
    Arrays
"""
x = "Priya"
print(x[3])
for p in x: # flows through values (the actual letters).
    print(p)
print("breathe")
'''
Preparation: Python first calculates len(x), which is 5.

Range Generation: range(5) creates a sequence of numbers: 0, 1, 2, 3, 4.

The First Pass: The loop assigns the first number (0) to i. Inside the loop, it looks at x[0] (the character at address 0), which is 'P', and prints it.

The Second Pass: The loop assigns the next number (1) to i. It looks at x[1], which is 'r', and prints it.

The Sequence Continues: This repeats for i = 2, i = 3, and i = 4.

Termination: Once it hits the end of the range (4), the loop realizes there are no more numbers in the sequence and stops.
'''
for i in range(len(x)): # lows through positions (the numbers 0 to 4), which you then use to find the values.
    print(x[i])
```

```python
# Reverse string
for i in range(len(x)-1, -1, -1): # len is used with
    print(x[i])
for let in x[::-1]:
    print(let)
```
